1. With energy and momentum conservations, one can get $$p=\frac{E_\gamma}{c}=\sqrt{2mE_\text{recoil}} \text{.}$$
So the ratio $$\frac{E_{\text{recoil}}}{E_\gamma}=\frac{E_\gamma}{2mc^2}<\frac{Q}{2mc^2} \text{.}$$
The Q-value of this process is $$Q=3*m(\alpha)*c^2-m(^{12}C)*c^2=7.275 \text{ MeV,}$$ which is much smaller as the rest energy $$mc^2=11200 \text{ MeV}.$$ So the ratio is close to 0, which means the recoil energy is negligible.

```python
delta_m_alpha = 2.42491587 #mass excess of alpha-particlein MeV
delta_m_carbon = 0 #mass excess of carbon 12 in MeV

Q = 3 * delta_m_alpha - delta_m_carbon
print("Q-Value:", Q)
```

2. (a) Parent Nuclide 1: $$\frac{dY_1}{dt}=-\lambda_1 Y_1 \text{,}$$
      Daughter Nuclide 2: $$\frac{dY_2}{dt}=\lambda_1 Y_1-\lambda_2 Y_2 \text{,}$$
      Stable Nuclide 3: $$\frac{dY_3}{dt}=\lambda_2 Y_2 \text{.}$$
   (b) Parent Nuclide 1: $$Y_1(t)=Y_0 e^{-\lambda_1 t}\approx Y_0(1-\lambda_1 t)\text{,}$$
      Daughter Nuclide 2: $$Y_2(t)=\frac{\lambda_1 Y_0}{\lambda_2-\lambda_1}(e^{-\lambda_1 t}-e^{-\lambda_2 t})\approx Y_0\lambda_1 t\text{,}$$
      Stable Nuclide 3: $$Y_3(t)=Y_0(1-\frac{\lambda_2e^{-\lambda_1 t}-\lambda_1e^{-\lambda_2 t}}{\lambda_2-\lambda_1})\approx 0\text{.}$$
   (c) Intermediate nuclide $Y_i$ (quasi-stable): $$\frac{dY_i}{dt}=\lambda_{i-1} Y_{i-1}-\lambda_i Y_i\approx 0\text{.}$$
      So that $$\lambda_i Y_i \approx \lambda_1 Y_1\text{.}$$

3. (a) Radiation dominated: phonon gas, follows the $P=\frac{1}{3}u=bT^4$. \
   (b) Non-degenerate ideal gas: ideal gas, follows the state-equation $pM=\rho RT$. \
   (c) Fully degenerate no-relativistic: Fermi-Dirac statistics, follows the $P=\frac{2}{5}\varepsilon_F n\propto
   \rho^{5/3}$.
   (d) Fully degenerate ultrarelativistic: degenerate effect dominates.

```python
# Define temperature (T) range
T = np.arange(0, 1e9, 1e6)

# Define boundary-functions
rho1 = 3e-23 * T**3
rho2 = 2.4e-8 * T**(3/2)

# Set up the plot
plt.figure(figsize=(10, 8))

# Plot boundaries of regions
plt.fill_between(T, rho1, 0, color="orange", alpha=0.3, label="Radiation Dominated") # Radiation Dominated

plt.fill_between(T, rho1, rho2, color="blue", alpha=0.3, label="Non-Degenerate Ideal Gas") # Non-degenerate Ideal Gas

plt.fill_between(T, rho2, 1.73e6, color="green", alpha=0.3, label="Fully Degenerate Non-Relativistic") # Fully Degenerate, Non-Relativistic

plt.fill_between(T, 1.73e6, 2e6, color="purple", alpha=0.3, label="Fully Degenerate Ultra-Relativistic") # Fully Degenerate, Ultra-Relativistic

# Plot of some systems
plt.plot(10e7, 100, 'ro', markersize=8, label="Stellar Core") # Stellar core

plt.plot(1e5, 1e6, 'bo', markersize=8, label="White Dwarf") # White dwarf

plt.plot(300, 10, 'go', markersize=8, label="Metallic Solid (Room temperatur)") # Normal metallic solid at room temperature and pressure

# Labels and legend
plt.xlabel("Temperature $T$ [K]")
plt.ylabel("Density $\\rho$ [g/cm]")
plt.title("$\\rho-T$ Diagram for a Fermion Gas")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```
